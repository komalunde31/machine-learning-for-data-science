```python
#create multiple plots in single canvas
fig, axes = plt.subplots(2,2,  figsize=(10,10))
axes[0,0].set_title("Sepal Length")
axes[0,0].hist(df["sepal_len"], bins=7)

axes[0,1].set_title("Sepal Width")
axes[0,1].hist(df["sepal_wid"], bins=7)

axes[1,0].set_title("Petal Length")
axes[1,0].hist(df["petal_len"], bins=7)

axes[1,1].set_title("Petal Width")
axes[1,1].hist(df["petal_wid"], bins=7)
```

```python
#ploting multiple plots together
plot = sns.FacetGrid(df,hue="species")
plot.map(sns.distplot, "sepal_len").add_legend()

plot = sns.FacetGrid(df,hue="species")
plot.map(sns.distplot, "sepal_wid").add_legend()

plot = sns.FacetGrid(df,hue="species")
plot.map(sns.distplot, "petal_len").add_legend()

plot = sns.FacetGrid(df,hue="species")
plot.map(sns.distplot, "petal_wid").add_legend()
```

```python
sns.FacetGrid(df, hue="species" , size=5).map(plt.scatter,"sepal_len","sepal_wid").add_legend()
```

```python
#alpha - color strength based on alpha value
sns.relplot(x="sepal_len", y="petal_len", hue = "species",
           data= df, kind ="scatter", sizes=(5,60), alpha=0.6)
```

```python
#regression line- trend/relationship between two variable
sns.regplot(x="sepal_len", y="petal_len", 
           data= df)
```
