```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
#read data set
df = pd.read_csv("train.csv")
df.head()
```

```python
#determine the standard deviation, mean, lowest value, max value and four quartile percentiles.
df.describe()
```

```python
X = df[['trip_distance','rate_code','payment_type','fare_amount','extra','mta_tax','tip_amount','tolls_amount','imp_surcharge','total_amount','month','day_of_week','hour_of_day','trip_duration']]
X.head()
```

```python
y = df['calculated_total_amount']
y.head()
```

```python
#add a dummy column to match no. of columns with beta values
X = sm.add_constant(X)
X.head()
```

```python
#train a model-Ordinary Least Square
model1 = sm.OLS(y, X).fit()
```

```python
# displaying the model
print(model1.summary())
```
