```python
from sklearn.model_selection import train_test_split
X = df.loc[:, df.columns != 'calculated_total_amount']
y = df['calculated_total_amount']
X_train , X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 1)
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression().fit(X, y)
```

```python
model1.coef_
```

```python
model1.score(X_test, y_test)
```

```python
pred = model1.predict(X_test)
pred
```

```python
df1 = X_test.copy()
df1['calculated_total_amount'] = y_test
df1['predicted_total_amount'] = pred
df1
```

```python
df1['Square_Error'] = (df1['calculated_total_amount'] - df1['predicted_total_amount'])**2
df1
```

```python
RMSE = (np.sum(df1['Square_Error']))**0.5
RMSE
```
