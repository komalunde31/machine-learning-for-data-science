# My First Multiple Linear Regression Model using StatsModel

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_csv("auto_mpg.csv")
df.head()
```

```python
X = df[['cyl','disp','hp','weight','acc','year','origin']]
X.head()
```

```python
y = df['mpg']
y.head()
```

```python
X = sm.add_constant(X)
X.head()
```

```python
model1 = sm.OLS(y, X).fit()
```

```python
print(model1.summary())
```
